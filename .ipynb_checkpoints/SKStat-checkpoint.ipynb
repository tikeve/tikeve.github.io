```python
hist_num = 6

'''
    Making all input tables standart format. Only numer columns. From left to right. No lists, columns for DGW added
'''
df = TxGA1920#no_lists(Team_xG_Ad.sort_index()[Team_xG_Ad.columns[:3:-1]])
df_answers = TxG1920#no_lists(Team_xG.sort_index()[Team_xG.columns[:3:-1]])
df_opp = Todc1920
threatAllowedAv = TOxG1920.mean(axis=1).mean()

X3 = pd.DataFrame()
X3_opponents = []
X3_answers = []
col = []

for i in range(hist_num, 0, -1):
    col.append(f'GW-{i}')

for i in range(0,len(df.columns)-hist_num):
    #print(i+1)
    next_tour2add = df[df.columns[i:i+6]].copy()#no_lists(df.sort_index())[df.columns[i:i+6]]
    next_tour2add.columns=col#['GW-6','GW-5','GW-4','GW-3','GW-2','GW-1']
    next_tour2add['Opponent'] = (df_opp[df_opp.columns[hist_num+i]]*threatAllowedAv)#.copy()
    next_tour2add['GW'] = df_answers[df_answers.columns[hist_num+i]]#.copy()#.tolist()
    X3 = X3.append(next_tour2add, ignore_index=True, sort=False)
    
    #X3_opponents.extend(Todc[df.columns[hist_num+i]]*threatAllowedAv)
    
    #X3_answers.extend(df_answers[df_answers.columns[hist_num+i]].tolist())
#X3['Opponent'] = X3_opponents
#X3['GW'] = X3_answers
#print(X3_opponents)
X3 = X3.dropna()
X = X3[X3.columns[:-1]]
Y = X3['GW']
reg = LinearRegression().fit(X, Y)
display(reg.score(X, Y))
display(reg.coef_)
```
