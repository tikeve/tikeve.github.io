```python
import sklearn
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
from pathlib import Path
from Brr_functions import no_lists
import json
```

```python
with open(Path('mid/FPL/Team_xG.txt'), 'r') as file:
    Team_xG = pd.DataFrame(json.loads(file.read()))
    Team_xG.index = pd.to_numeric(Team_xG.index)
    Team_xG = Team_xG.sort_index()
    
with open(Path('mid/FPL/Team_Opponent_xG.txt'), 'r') as file:
    Team_Opponent_xG = pd.DataFrame(json.loads(file.read()))
    Team_Opponent_xG.index = pd.to_numeric(Team_Opponent_xG.index)
    Team_Opponent_xG = Team_Opponent_xG.sort_index()
    
with open(Path('mid/FPL/Team_oppdefence_coeff.txt'), 'r') as file:
    Team_oppdefence_coeff = pd.DataFrame(json.loads(file.read()))
    Team_oppdefence_coeff.index = pd.to_numeric(Team_oppdefence_coeff.index)
    Team_oppdefence_coeff = Team_oppdefence_coeff.sort_index()
    
threatAllowedAv = no_lists(Team_Opponent_xG[Team_Opponent_xG.columns[4:]]).mean(axis=1).mean()
lastGW = len(Team_xG.columns[4:])
    
```

```python
df = no_lists(Team_xG.sort_index())#sort_index doesn't matter as it adds column to suited index anyway


X = df[df.columns[5:-1]].copy()
X['opp_def'] = Team_oppdefence_coeff['GW8']*threatAllowedAv
Y = df['GW8']
reg = LinearRegression().fit(X, Y)
display(reg.score(X, Y))
display(reg.coef_)

X = df[df.columns[5:-1]].copy()
X['sum'] = X.sum(axis=1)/6
X['opp_def'] = Team_oppdefence_coeff['GW8']*threatAllowedAv
X = X[X.columns[-2:]]
display(X)
Y = df['GW8']
reg = LinearRegression().fit(X, Y)
display(reg.score(X, Y))
display(reg.coef_)
```
