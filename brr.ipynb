```python
''' Main file

    Main module. Runs program(), does some testing, sends mistakes on mail and telegram
    
    
    Uses 3 modules: inputFPL, inputUnderstat, Source
    
    inputFPL downloads data from https://fantasy.premierleague.com/
    inputUnderstat downloads data from https://understat.com/
    Source converts data to suitable pandas tables correctness by running source.test function.
    
    
    Uses no data

'''


# Основной файл.
#Скачивает данные через модули inputfpl1.py и inputUnderstat
#Преобразовывет их в таблицы таблицы через конструктор класса Source
#Ф-ция program делает все необходомое, нужна чтобы использовать её по желанию с почтой(telegram) и без.
import yagmail
import requests
import os
from dotenv import load_dotenv
from time import time
import runpy
import sys
from traceback import format_exc

from Source import Source
import constti

# Скачаивает и записывает в таблицы все данные. Выводит в файоы папки input и output/Source
def program():
    start_program = time()
    
    sys.argv = ['', '*']
    #Loading Data from FPL and Understat
    runpy.run_module('inputFPL', run_name='smth')
    runpy.run_module('inputUnderstat', run_name='smth')

    
    #Берет данные fantasy.premierleague.com(FPl) и делает из них финальные таблицы. ma_num - gameweeks to average
    FPL = Source('FPL', ma_num=7)
    #Takes data from understat.com(Understat) and builds table out of them. ma_num - gameweeks to average
    Understat = Source('Understat', ma_num=7)
    #Tests for the Final tables
    FPL.test()
    Understat.test()
    
    print(f'The whole program working time is  {time() - start_program}')
    #display(FPL.PlayerThreatAd)
    print('End')

#send mail or not
with_mail = 1

# Если отправлять ошибки на почту
if with_mail:
    try:
        program()
    except Exception as e:
        
        err = format_exc()
        
        #Reading passwords from .env
        load_dotenv()
        telegram_API = os.getenv('telegram')
        mail_pass = os.getenv('imonyapa_mail')
        
        #Sending Error message to mail
        FROM = 'imonyapa@gmail.com'
        TO = 'constti@gmail.com'
        SUBJECT = '1stpy'
        TEXT=f'{e} \n\n\n {err}'
        yag = yagmail.SMTP(FROM, mail_pass)
        yag.send(TO, SUBJECT, TEXT)
        
        #Sending Error message with telegram
        url = f'https://api.telegram.org/bot1182664319:{telegram_API}/sendMessage?chat_id=108651591&text='+TEXT

        http_proxy  = "http://10.10.1.10:3128"
        https_proxy = "https://51.178.59.37:3128"
        ftp_proxy   = "ftp://10.10.1.10:3128"

        proxyDict = { 
                      "http"  : http_proxy, 
                      "https" : https_proxy, 
                      "ftp"   : ftp_proxy
                    }
        
        #Proxy is dead(((
        p=requests.get(url)#, proxies=proxyDict)
        
        print(TEXT)
# Без отправки ошибок на почту        
else: program()
```
