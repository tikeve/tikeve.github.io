```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../mid/FPL/TxG.csv')
Home = pd.read_csv('../mid/Team_home.csv')
Defence = pd.read_csv('../mid/FPL/TOxG.csv')
Opponents = pd.read_csv('../in/Team_opponent_team.csv')
```

```python
def no_doubles(df):
    X = pd.DataFrame()
    for i in df.index:
        X = X.append(pd.DataFrame(df.loc[i,:].dropna().reset_index(drop=True)).T)
    return X


def to_table(df, lag, is_y=False):
    X = pd.DataFrame()
    limit = 39
    if is_y:
        limit = 38
        y=[]
    for i in range(lag, limit):
        temp = df.iloc[:, i-lag:i]
        temp = temp.rename(columns=dict(zip(temp.columns, range(len(temp.columns)))))
        X = X.append(temp).reset_index(drop=True)
        if is_y:
            y = y + (list(df.iloc[:, i]))
    if is_y:
        return X, y
    return X
    
X = no_doubles(df)
X, y = to_table(X, 5, is_y=True)

H = no_doubles(Home)
HA, H = to_table(H, 5, is_y=True)
#X = X.join(HA, lsuffix='lag', rsuffix='HA')
X['Home'] = H


O = no_doubles(Opponents)


D_coef = no_doubles(Defence)
D_coef = D_coef.mean(1)
D_dict = dict(zip(list(range(20)), D_coef))
DD = O.applymap(lambda x: D_dict[x-1])
DD, D = to_table(DD, 5, is_y=True)
X['Op def'] = np.array(D)

print(len(X))
len(H)
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from catboost import CatBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor

seed = 1111
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = seed)
```

```python
model_lr = Ridge(alpha=0)
model_lr.fit(X_train, y_train)

RMSE_train = np.sqrt(mean_squared_error(y_train, model_lr.predict(X_train)))
RMSE_test = np.sqrt(mean_squared_error(y_test, model_lr.predict(X_test)))
display(model_lr.coef_)
print(f'Train RMSE: {RMSE_train}')
print(f'Test RMSE: {RMSE_test}')
```

```python
model_rf = RandomForestRegressor(n_estimators=100)
model_rf.fit(X_train, y_train)

RMSE_train = np.sqrt(mean_squared_error(y_train, model_rf.predict(X_train)))
RMSE_test = np.sqrt(mean_squared_error(y_test, model_rf.predict(X_test)))
print(f'Train RMSE: {RMSE_train}')
print(f'Test RMSE: {RMSE_test}')
```

```python
model_cat = CatBoostRegressor(verbose=100, random_state=seed, l2_leaf_reg=111)
model_cat.fit(X_train, y_train)

RMSE_train = np.sqrt(mean_squared_error(y_train, model_cat.predict(X_train)))
RMSE_test = np.sqrt(mean_squared_error(y_test, model_cat.predict(X_test)))
print(f'Train RMSE: {RMSE_train}')
print(f'Test RMSE: {RMSE_test}')
```

```python
model_lgbm = LGBMRegressor(random_state=seed,reg_lambda=111)
model_lgbm.fit(X_train, y_train)

RMSE_train = np.sqrt(mean_squared_error(y_train, model_lgbm.predict(X_train)))
RMSE_test = np.sqrt(mean_squared_error(y_test, model_lgbm.predict(X_test)))
print(f'Train RMSE: {RMSE_train}')
print(f'Test RMSE: {RMSE_test}')
```
