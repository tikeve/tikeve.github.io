```python
from numpy.random import binomial
#import matplotib as plt
def count_frequencies (X, relative=False):
    a = dict()
    for i in X:
        if i in a.keys():
            a[i] += 1
        else:
            a[i] = 1
            
    if relative:
        for i in a.keys():
            a[i] = a[i]/len(X)
    return a
n = 10
p=0.3
sample = binomial(n,p, size=10000)
freqs = count_frequencies(sample, relative=True)

freqs
```

```python
from scipy.stats import binom
X = binom(n, p)
X.pmf(1)
```
