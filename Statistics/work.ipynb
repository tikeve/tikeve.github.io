```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import ttest_1samp
```

```python
sample = [1,3,5,2,4,1,4,4,5,6,5,3,1,3,3,4,20]
mu0 = 2
t_obs = ttest_1samp(sample, mu0)[0]
t_obs
```

```python
np.std(sample, ddof=1)
```

```python
from scipy.stats import t
x= np.linspace(-5,5)
T = t(df=len(sample) - 1)
plt.plot(x,T.pdf(x))
x_left = np.linspace(-5,-t_obs, 200)
x_right = np.linspace(t_obs,5,200)
plt.fill_between(x_left, T.pdf(x_left))
plt.fill_between(x_right, T.pdf(x_right))
```

```python
T.cdf(-t_obs)*2
```
