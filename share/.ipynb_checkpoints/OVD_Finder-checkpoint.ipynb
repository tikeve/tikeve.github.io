```python
#Пересмотреть пересечения
#Заново просеить Яндекс, предварительно сохранив старые данные
#Попробовать из этого сделать вариант для волонтеров
#Загрузить в googleDoc
#Написать описание задания

#Понять как дешифровать КЛАДР. Добавить кладр в файл. Что такое ФИАС.(сравнить коодринаты из адресов и КЛАДРа)
#openstreetmap (гис??) Сделать аналогичную таблицу.

import pandas as pd
import requests
import math
from bs4 import BeautifulSoup
import numpy as np

import requests
from collections import OrderedDict

def getAddress(i,new):
    #Возвращает адресс по номеру строчки в таблице
    sub = StrNoNANs(new.iat[i,1])
    distr = StrNoNANs(new.iat[i,2])
    city = StrNoNANs(new.iat[i,3])
    street =StrNoNANs(new.iat[i,4])
    house = StrNoNANs(new.iat[i,5])
    #return str(new.iat[i,1]) + ', ' + str(new.iat[i,2]) + ', ' + str(new.iat[i,3]) + ', ' + str(new.iat[i,4]) + ', ' + str(new.iat[i,5])
    return sub + ', ' + distr + ', ' + city + ', ' + street + ', ' + house

def StrNoNANs(t):
    t = str(t)
    if t == 'nan':
        return ''
    return t

def LatLong(adress):
    #Возвращает широту и долготу по адресу
    url = 'https://geocode-maps.yandex.ru/1.x/?geocode= ' + str(adress)
    try:
        page = requests.get(url)
        ttex = BeautifulSoup(page.text, 'html.parser')
        return [float(ttex.point.pos.string.split()[0]), float(ttex.point.pos.string.split()[1])]
    except:
        return[0, 0]

def LongLat2Address(LL):
    #Возвращает адрес по долготе и широте
    if type(LL)!=str:
        LLstring=str(LL[0])+','+str(LL[1])
    else:
        LLstring=LL[1:-1]
    url = 'https://geocode-maps.yandex.ru/1.x/?geocode=' + LLstring
    try:
        page = requests.get(url)
        ttex = BeautifulSoup(page.text, 'html.parser')
        return ttex('text')[0].string
    except:
        return None
    
def LatLongDist(o1, o2):
    #Расчёт расстояния в метрах по широте и долготе
    
    #obj1 = [o1[0],o1[1]]
    #obj2 = [o2[0],o2[1]]
    if type(o1) == str:
        o1 = [float(o1[1:-1].split(',')[0]),float(o1[1:-1].split(',')[1])]
    if type(o2) == str:
        o2 = [float(o2[1:-1].split(',')[0]),float(o2[1:-1].split(',')[1])]
    
    obj1 = [o1[0],o1[1]]
    obj2 = [o2[0],o2[1]]
    
    obj1[0]= obj1[0]*math.pi/180
    obj1[1]= obj1[1]*math.pi/180
    obj2[0]= obj2[0]*math.pi/180
    obj2[1]= obj2[1]*math.pi/180
    R = 6371000
    cosD = math.sin(obj1[0])*math.sin(obj2[0]) + math.cos(obj1[0])*math.cos(obj2[0])*math.cos(obj1[1] - obj2[1])
    #Для того, чтобы ошибки вычислений не приводили к ошибке
    if (cosD > 1) & (cosD < 1.01):
        cosD = 1
    dist = R*math.acos(cosD)
    return dist

def getPolice(ll_centr, spn):
    #Возвращает JSON со списком отделов полиции с центром ll_centr и диаметрами spn (широта-долгота)
    apikey = '9b2655e9-a245-4d79-888a-dff4bb85f53e'
    zapros = 'Полиция'
    lang = 'ru_RU'
    ll_centr = str(ll_centr[0])+','+str(ll_centr[1])
    spn = str(spn[0])+','+str(spn[1])
    #ll_centr = '37.618920,55.756994'
    #spn = '8,2'# '0.552069,0.400552'
    url='https://search-maps.yandex.ru/v1/?apikey='+apikey+'&text='+zapros+'&lang='+lang+'&ll='+ll_centr+'&spn='+spn+'&rspn=1&results=500'
    return requests.get(url)

def page2Table(page):
    #Преобразует API yandex в таблицу
    js = page.json()
    YOVD=pd.DataFrame(columns=['Name', 'Address', 'LngLat', 'Phones', 'Site'])

    for i in range(len(js['features'])):
        #IsRussia1=False
        #IsRussia2=False
        #IsRussia3=False
        name = js['features'][i]['properties'].get('name')
        address = js['features'][i]['properties']['CompanyMetaData'].get('address')
        #if address!=None:
        #    IsRussia1 = (address.split(',')[0]=='Россия')
        LL = js['features'][i]['geometry']['coordinates']
        allPhones = js['features'][i]['properties']['CompanyMetaData'].get('Phones')
        if allPhones:
            Phones = list(OrderedDict.fromkeys([ph['formatted'] for ph in allPhones]))
            #IsRussia2 = (allPhones[0].get('country')=='7')
        else:
            Phones = None
        Site = js['features'][i]['properties']['CompanyMetaData'].get('url')
        #if Site!=None:
        #    IsRussia3=(Site[-3:] in {'ru/', 'рф/'})
        
        #if IsRussia1|IsRussia2|IsRussia3:
        YOVD.loc[i]=[name, address, LL, Phones, Site]

    return [YOVD, len(js['features'])]

def subRight(t):
    if t=='РД':
        return 'Дагестан'
    if t=='КЧР':
        return 'Карачаево-Черкессия'
    if t=='МВД по Республике Тыва':
        return 'Тыва'
    if t=='Кракснодарский край':
        return 'Краснодарский край'
    if t=='Ставрпольский край':
        return 'Ставропольский край'
    if t=='Сах. Обл':
        return 'Сахалин'
    if t=='ЗАТО Озерный':
        return ''
    if t=='УР':
        return 'Удмуртия'
    return t


def distrRight(t):
    if t=='Полазненское СП':
        return 'Добрянский район'
    if t=='Лискинский муниц. район Воронежской области':
        return 'Лискинский муниц. район'
    if t=='Подгоренский муниц. район Воронежской области':
        return 'Подгоренский муниц. район'
    if t=='ВМЗ, филиал ФГУП «ГКНПЦ им. В.М. Хруничева», ОАО «Концерн «Созвездие», ОАО «КБХА», ИК ОАО «КБХА» ':
        return ''
    if t=='Падунский округ г. Братска':
        return 'Падунский округ'
    if t=='г.Энгельс Энгельсского муниципального района':
        return 'г.Энгельс'
    if t=='ЗАТО Озерный':
        return ''
    return t

def streetRight(t):
    if t=='65-66 км МКАД':
        return ''
    return t

def streetHouse(t):
    if t=='б/н':
        return ''
    return t

def InList(t):
    answer = False
    t = t.lower()
    SetAccepted = {1000, t.find(' оп '), t.find('отдел полиции'), t.find('отделение полиции'), t.find('отд мвд'),
                   t.find('омвд'),t.find('отделение мвд'), t.find('оемвд'),
                   t.find('отд-е'), t.find('Отдел МВД'.lower()), t.find('Отдел министерства внутренних дел'.lower()), 
                   t.find('Отделение министерства внутренних дел'.lower()),
                   t.find(' ОВД'.lower()), t.find('ОВД '.lower()), t.find('РОВД'.lower()), t.find(' ПОП '.lower()),
                  t.find('Отделений полиции'.lower()), t.find('поселковое отделение'), t.find('Отел МВД'.lower()),
                   t.find('Отделполиции'.lower()), t.find('ОтделМВД'.lower()), t.find('отдел полици'.lower()),
                   t.find('отдедение полиции'),
                  t.find('Отдела МВД России по району Кузьминки г.Москвы'.lower()),
                  t.find('Ясенево'.lower()),
                  t.find('Южное Бутово'.lower()),
                   t.find('Теплый Стан'.lower()),
                  t.find('Черемушки'.lower()),
                  t.find('СеверноеБутово'.lower()),
                  t.find('Обручевский'.lower()),
                  t.find('Ломоносовский'.lower()),
                  t.find('Котловка'.lower()),
                  t.find('Коньково'.lower()),
                  t.find('Зюзино'.lower()),
                  t.find('Гагаринский'.lower()),
                  t.find('Академически'.lower()),
                  t.find('Царицыно'.lower()),
                    t.find('Орехово – Борисово Южное'.lower()),
                   t.find('Орехово - БорисовоСеверное'.lower()),
                  t.find('Нагорный'.lower()),
                  t.find('Нагатино – Садовники'.lower()),
                   t.find('Нагатинский Затон'.lower()),
                  t.find('Москворечье - Сабурово'.lower()),
                  t.find('Донской'.lower()),
                  t.find('Даниловский'.lower()),
                  t.find('Зябликово'.lower()),
                  t.find('Братеево'.lower()),
                  t.find('Бирюлево Западное'.lower()),
                  t.find('Бирюлево Восточное'.lower()),
                  t.find('Ярославский'.lower()),
                  t.find('Северный'.lower()),
                  t.find('Свиблово'.lower()),
                  t.find('Ростокино'.lower()),
                  t.find('Отрадное'.lower()),
                  t.find('Останкинский'.lower()),
                  t.find('Южное Медведково'.lower()),
                  t.find('Северное Медведково'.lower()),
                  t.find('Марьина Роща'.lower()),
                  t.find('Марфино'.lower()),
                  t.find('Лосиноостровский'.lower()),
                  t.find('Лианозово'.lower()),
                  t.find('Бутырский'.lower()),
                  t.find('Бибирево'.lower()),
                  t.find('Бабушкинский'.lower()),
                  t.find('Алтуфьевский'.lower()),
                  t.find('Алексеевский'.lower()),
                  t.find('Чертаново Северное'.lower()),
                  t.find('Чертаново Центральное'.lower()),
                    t.find('Чертаново Южное'.lower())}
    
    SetRejected = {1000,t.find('пункт'), t.find('участков'), t.find('линейн'), t.find('ЛОВД'.lower()), t.find('при ОВД'.lower()),
                   t.find('Комплекс'.lower()), t.find('упп'), t.find('обслуживан'),
                    t.find('МОВД'.lower()), t.find('МОМВД'.lower()), t.find('пульт'.lower()),
                  t.find('межмуниципальный отдел'.lower()), t.find('дежурн'.lower())}
    
    SetAccepted.discard(-1)
    SetRejected.discard(-1)
    
    SetStrictRejected = {'метро', 'экономи', 'пдн', "Цвснп", 'миграц',
                         'Следств', 'спецприем', 'наркотик', 'несовершеннолет', 'Комплекс',
                        'ГИБДД', 'содержани', ' ЛОП ', ' ПП ','ФГУП', ' ОВО ', 'отделение полиции МГУ' }
    SetExceptions = {'Метрогородок', 'Отдел полиции по обслуживанию Ловозерского района', 'Отделение полиции по Урупскому району',
                    'по обслуживанию а. Псыж'} 
    
    if min(SetAccepted) < min(SetRejected):
        #print(min(SetAccepted), min(SetRejected))
        answer = True
    
    if (t[:8].lower()=='межмуниц')|(t[:4].lower()=='ово ')|(t[:4].lower()=='лоп ')|(t[:4].lower()=='эко '):
        #print(1)
        answer = False
    if (t[:3].lower()=='оп ')|(t[:6].lower()=='отдел ')|(t[:4].lower()=='поп ')|(t[-4:].lower()==' поп')|(t[:5].lower()=='о мвд'):
        #print(2)
        answer = True
    if (t[:9].lower()=='отделение ')|(t[:6].lower()=='отдел ')|(t[:3].lower()=='оп№')|(t[:8].lower()=='отдление'):
        #print(3)
        answer = True
    if (t[:16].lower()=='отделение полции')|(t[-3:].lower()==' оп')|(t[:3].lower()=='оп.')|(t[:3].lower()=='оп-'):
        #print(4)
        answer = True
    if (t.lower() == 'отделение полиции')|(t.lower() == 'отдел полиции'):
        #print(5)
        answer = False
    if (t[:3].lower()=='пп ')|(t[-3:].lower()==' пп')|(t[:4].lower()=='увд '):
        answer = False
    for a in SetStrictRejected:
        if a.lower() in t.lower():
            answer = False
    for a in SetExceptions:
        if a.lower() in t.lower():
            answer = True
    return answer

def searchArea(ll_centr, spn):
    ToAdd,num = page2Table(getPolice(ll_centr, spn))
    if num==500:
        #problems.append(ll_centr)
        T1 = searchArea([ll_centr[0]-spn[0]/4.0, ll_centr[1]-spn[1]/4.0], [spn[0]/2.0,spn[1]/2.0])
        T2 = searchArea([ll_centr[0]+spn[0]/4.0, ll_centr[1]-spn[1]/4.0], [spn[0]/2.0,spn[1]/2.0])
        T3 = searchArea([ll_centr[0]-spn[0]/4.0, ll_centr[1]+spn[1]/4.0], [spn[0]/2.0,spn[1]/2.0])
        T4 = searchArea([ll_centr[0]+spn[0]/4.0, ll_centr[1]+spn[1]/4.0], [spn[0]/2.0,spn[1]/2.0])
        return pd.concat([T1,T2,T3,T4]).drop_duplicates(subset=['Address','Name','Site']).reset_index(drop=True)
    else:
        return ToAdd
def near(x):
    if x<100:
        return '!!!!!!!!!!!!'
    else:
        if x<1000:
            return '+++'
        else:
            return ''
def CheckCloseOVD(t):
    SubSet=pd.DataFrame(columns=['Name','Source','Address','Phones','Lng', 'Lat','i'])
    k,l=0,0
    for i in range(len(t)):
        a=0
        obj1 = [t.iat[i,5], t.iat[i, 4]]
        for j in range(i+1,len(t)):
            obj2 = [t.iat[j,5], t.iat[j, 4]]
            if LatLongDist(obj1,obj2)<distance:
                #print(LatLongDist(obj1,obj2))
                
                SubSet.loc[l]=t.iloc[i]
                SubSet.iat[l,6]=i+2
                SubSet.loc[l+1]=t.iloc[j]
                SubSet.iat[l+1,6]=j+2
                l+=2
                break
    return SubSet
```
